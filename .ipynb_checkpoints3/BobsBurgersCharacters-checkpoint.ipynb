```python
#Import packages
import requests
import pandas as pd
```

```python
#Read in API
r = requests.get("https://bobsburgers-api.herokuapp.com/characters/?limit=9&skip=168")
```

```python
#Convert API to JSON
json = r.json()
print(json)
```

```python
#Create DataFrame
df = pd.DataFrame(json)
print(df)
```

```python
#Find and print two descriptive statistics

#Description statistics for character column
character_stats = df['name'].describe()
print(character_stats)

#How many characters listed?
characters = df['name'].count()
print(characters)
```

```python
#Write a query in Pandas

#Viewing only female characters
female = df.query("gender == 'Female'")
print(female)
```

```python
#Select second and third columns of DataFrame
columns = df.iloc[:, 1:3]
print(columns)
```

```python
#Select first four rows of DataFrame
rows = df[0:4]
print(rows)
```
